## 실습 : Zipf's Law

제공된 품사부착 말뭉치를 분석하여 빈도수가 큰 상위 100개의 형태소를 아래와 같은 형태로 출력하세요 (10wja)

- 사용언어 : 무관
- 입력 형태 : 어절 ₩(spacing unit) 형태소/품사+형태소/품사 +. .. (외울 필요 x)
- 출력 형태 : 형태소/품사, 빈도수, 순위*빈도수 (=c)
- 제출 방법 : 소스코드와 출력 결과를 포함 (ipynb 추천)

```python
file_path = "/gdrive/MyDrive/NLP/week1/train_data_after_10000_sent.txt"

with open(file_path, "r", encoding="cp949") as inFile:
  lines = inFile.readlines()
```

```python
lines
```

```python
import operator

morpheme_frequency_dict = {}

for line in lines:
    words = line.replace('\n', '').split('\t')
    morphemes = words[1].split('+')
    
    for morpheme in morphemes:
        if morpheme in morpheme_frequency_dict:
            morpheme_frequency_dict[morpheme] += 1
        else:
            morpheme_frequency_dict[morpheme] = 1

mor_fre_dict = sorted(morpheme_frequency_dict.items(), key=operator.itemgetter(1), reverse=True)[:100]

rank = 1

for mor, fre in mor_fre_dict:
    print(mor, fre, fre * rank) 
    rank += 1
```
